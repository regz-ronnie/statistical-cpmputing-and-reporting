What is the Mean value of each column against each 'Weather Condition ?

```python
data.groupby('Weather')[['Temp_C','Dew Point Temp_C','Rel Hum_%','Wind Speed_km/h','Visibility_km','Press_kPa']].mean()
```

Daily temperature and humodity pattern

```python
data['Date/Time']=pd.to_datetime(data['Date/Time']) # converting into date/time format
daily_avg = data.groupby(data['Date/Time'].dt.date)[['Temp_C', 'Rel Hum_%']].mean()
plt.figure(figsize=(12, 6))
plt.plot(daily_avg['Temp_C'], label='Temp_C')
plt.plot(daily_avg['Rel Hum_%'], label='Rel Hum_%')
plt.xlabel('Date/Time')
plt.ylabel('Average Value')
plt.title('Daily Temperature and Humidity Patterns')
plt.legend()
plt.show()
```

What is the average temperature and humidity at different times of the day?

```python
data['hour']=data['Date/Time'].dt.hour
day_avg = data.groupby('hour')[['Temp_C', 'Rel Hum_%','Wind Speed_km/h']].mean()
plt.figure(figsize=(12, 6))
plt.plot(day_avg['Temp_C'], label='Temp_C')
plt.plot(day_avg['Rel Hum_%'], label='Rel Hum_%')
plt.plot(day_avg['Wind Speed_km/h'], label='Wind Speed_km/h')
plt.xlabel('hour')
plt.ylabel('Average Value')
plt.title('Daily Temperature,Humidity and Wind Speed Patterns')
plt.legend()
plt.show()
```

Impact of Wind Speed on Temperature and Humidity

```python
data['Wind_Category']=pd.cut(data['Wind Speed_km/h'],bins=[0,30,60,90],labels=['low','medium','high'])
wind_effect= data.groupby('Wind_Category')[['Temp_C','Rel Hum_%']].mean()
wind_effect
```

Monthly Visibility Analysis

```python
data['month']=data['Date/Time'].dt.month
monthly_visibility = data.groupby('month')['Visibility_km'].mean()

plt.figure(figsize=(10, 6))
plt.plot(monthly_visibility, marker='o', color='purple')
plt.xlabel('month')
plt.ylabel('Average Visibility')
plt.title('Average Monthly Visibility')
plt.show()
```
